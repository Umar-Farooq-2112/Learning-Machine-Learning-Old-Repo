```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("Position_Salaries.csv")
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.linear_model import LinearRegression
l_regressor = LinearRegression()
l_regressor.fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
p_regressor = PolynomialFeatures(degree = 4)
x_pol = p_regressor.fit_transform(x)
l_reg_2 = LinearRegression()
l_reg_2.fit(x_pol,y)
```

```python
plt.scatter(x,y,color = 'red')
plt.plot(x,l_regressor.predict(x),color='blue')
plt.title('Truth or Bluff(Linear Regression)')
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x,y,color = 'red')
plt.plot(x,l_reg_2.predict(p_regressor.fit_transform(x)),color='blue')
plt.title('Truth or Bluff(Li near Regression)')
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```
